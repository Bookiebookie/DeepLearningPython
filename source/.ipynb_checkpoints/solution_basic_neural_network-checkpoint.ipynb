The dataset we will use to train our first neural network is Fashion-MNIST dataset. It is a dataset of Zalando's  article images—consisting of a training set of 60,000 images and a test set of 10,000 images. Each images is a 28x28 grayscale image, associated with a label from 10 classes. 

To summarize, Fashion MNIST dataset includes:
- 60,000 training examples
- 10,000 testing examples
- 10 classes
- 28×28 grayscale/single channel images

The Fashion- MNIST dataset comes pre-loaded in tensorflow, in the form of a set of four Numpy arrays:

```python
# load dataset
from tensorflow.keras.datasets import fashion_mnist
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
```

```python
# The labels are an array of integers, in the range [0, 9]. 
# These correspond to the class of clothing the image represents:
labels = np.unique(test_labels)
print(labels)
```

## Network architecture
The problem we are trying to solve is to classify the images of fashion item into 10 categories (0-9). For that we need to first define our network architecture. We will use [tf.keras](https://www.tensorflow.org/api_docs/python/tf/keras) module to build and train our model. 'tf.keras' is the  Keras API integrated into TensorFlow 2. When Google released TensorFlow 2.0 in June 2019, they declared that Keras is now the official high-level API of TensorFlow for quick and easy model design and training. 

With the release of Keras 2.3.0, Francois Chollet, the creater of Keras [stated](https://github.com/keras-team/keras/releases/tag/2.3.0) that:
- this is the first release of multi-backend Keras that supports TensorFlow 2.0. It maintains compatibility with TensorFlow 1.14, 1.13, as well as Theano and CNTK.
- This release brings the API in sync with the tf.keras API as of TensorFlow 2.0. However note that it does not support most TensorFlow 2.0 features, in particular eager execution. If you need these features, use tf.keras.

- This is also the last major release of multi-backend Keras. Going forward, <b>we recommend that users consider switching their Keras code to tf.keras in TensorFlow 2.0.</b>

For that, We will use [Keras Sequential API](https://www.tensorflow.org/api_docs/python/tf/keras/Sequential) to define our model.

our network consists of sequence of following layers:
- input layer
- two dense layers (also called fully-connected layer)

The first layer in our model is input layer which is entry point to network. The first Dense layer has 128 nodes (or neurons). The second (and last) layer is a 10-node softmax layer that returns an array of 10 probability scores that sum to 1.  Each score will be the probability that the current  image belongs to one of 10 fashion item classes.

---
<div class='student'>
   <b>student task </b>
create a simple neural network model with following architecture
 <ul>
        <li>input layer of dimensionality 784 </li>
        <li>one hidden layer with 128 neurons with ReLU activation function</li>
        <li>one output layer with 10 neurons with softmax activation function</li>
    </ul>

Hint
- https://www.tensorflow.org/api_docs/python/tf/keras/layers
---
</div>

```python
### Define the Network Architecture
def make_model():
    ''' function to make model '''
    model = tf.keras.Sequential(
        [
        tf.keras.layers.Input(shape=(784,)),
        tf.keras.layers.Dense(128, activation=tf.nn.relu),
        tf.keras.layers.Dense(10, activation=tf.nn.softmax)
    ])
    return model
```

You can call summary() method to see the size and shape of the network.

```python
# print the model summary
model = make_model()
model.summary()
```

Model summary printed above is useful for simple models. But it can be confusing for model that have multiple inputs or outputs. tf.keras module provides a function called 'plot_model' in it's  tf.keras.utils namespace to create a plot of the network neural network graph that can make more complex models easier to understand. This function will create plot of the network.

```python
tf.keras.utils.plot_model(
    model,
    to_file='model.png',show_shapes=True, show_layer_names=True)
```

Before we can train our model, we still need to define three more components as a part of compilation steps:
- an optimizer,
- a loss function
- metric to assess the model performance

---
<div class='student'>
    <b>student task:</b>
set up the compilation step with following parameter

- use RMSprop optimizer
- CategoricalCrossentropy as loss function, and 
- accuracy as metric
</div>
---

```python
### Compile the model
model.compile(optimizer=tf.keras.optimizers.RMSprop(),
              loss=tf.keras.losses.CategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

Before we process to train and evaluate the model, we will preprocess the data by reshaping it into the shape our model expect, and scaling it so that pixel values lies between 0-1. 
Since the first the layer in our model is input layer with dimensionality of 784 but our training images are stored in an array of shape (60000, 28, 28) of type uint8 with values in the [0, 255] interval. We need to transform it into a float32 array of shape (60000, 28 * 28) with values between 0 and 1.

---
<div class='student'>
    <b>student task</b>
    
- reshape and standarize the train and test data
</div>
---

```python
# reshape the training and test images
train_images = train_images.reshape((60000, 28 * 28))
test_images = test_images.reshape((10000, 28 * 28))

# Standardize data to have feature values between 0 and 1
train_images = train_images.astype('float32') / 255
test_images = test_images.astype('float32') / 255
```

Since we have used categorical_crossentropy loss function in our model, we also need to categorically encode the class labels. 

---
<div class='student'>
    <b>student task: </b>
    
- convert training and test class vector into categorial label 

hint:

https://www.tensorflow.org/api_docs/python/tf/keras/utils/to_categorical
</div>

```python
train_labels = tf.keras.utils.to_categorical(train_labels)
test_labels = tf.keras.utils.to_categorical(test_labels)
```

We are now ready to train our network, which is done via a call to the 'fit' method. 

---
<div class='student'>
    <b>student task:</b>
    
Train the model with    
- epoch = 10
- batch size = 128
- validation split = 0.2
</div>


```python
History = model.fit(train_images, train_labels,validation_split=0.2, epochs=10, batch_size=128)
```

The 'fit' method on Keras returns the History object. It records training metrics (loss and accuracy) for each epoch. The metrics are stored as dictionary in 'History.history' attribute of History object. We can use metrics collected in History object to create plots. The plots can be used to assess various information about the model.

- convergence speed of model over epoch
- whether the model has already converged in 
- whether the model is over-fitting or not and other 


```python
# summarize history for accuracy
plt.plot(History.history['acc'])
plt.plot(History.history['val_acc'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()
# summarize history for loss
plt.plot(History.history['loss'])
plt.plot(History.history['val_loss'])
plt.title('model loss')# Let's plot the 

plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.title('Traning accuracy vs Epoch');
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()
```

Now, let's check the accuracy of model in test set, which is done by calling 'evalute' method on model. 
Test the accuracy of model on test set. Use batch size of 128.

```python
test_loss, test_accuracy = model.evaluate(test_images, test_labels,   batch_size=128)
print('Accuracy on test dataset:', test_accuracy)
```

## predictions and exploring 

So, we have now trained the model and evaluated its performance on training and test set. Let's see how model predict on individual images on test set.

```python
# plot the predicted label and true label of image
def plot_image(i, true_labels, images):
    
    ''' function to plot prediction result along with class probability
    
    arguments :
     i : image index
     true_labels : true class label of test set images
     images : test set images
    
    '''
    
    
    
    img=(images[i])
    img=np.array([img])
    predictions_array=model.predict([img])
    true_label=true_labels[i].argmax()
    fig, (ax1,ax2) = plt.subplots(figsize=(6,9),ncols=2)
    ax1.grid(False)
    ax1.set_xticks([])
    ax1.set_yticks([])
  
    ax1.imshow(img.reshape(28,28), cmap=plt.cm.binary)

    predicted_label = np.argmax(predictions_array)
    if predicted_label == true_label:
        color = 'blue'
    else:
        color = 'red'
  
    ax1.set_xlabel("Predicted : {} {:2.0f}% \n True label:({})".format(class_names[predicted_label],
                                100*np.max(predictions_array),
                                class_names[true_label]),
                                color=color)
 
    ax2.barh(np.arange(10), predictions_array.squeeze())
    ax2.set_aspect(0.1)
    ax2.set_yticks(np.arange(10))
    ax2.set_yticklabels(class_names, size='small');
    ax2.set_title('Class Probability')
    ax2.set_xlim(0, 1.1)

    plt.tight_layout()
    
```

```python
plot_image(100,test_labels, test_images)
```
