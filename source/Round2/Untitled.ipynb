```python
w1 = 10
w2 = -10
x = np.arange(-10,10,0.1)

y1 = 20 + w1*x
y2 = 20 + w2*x

g1 = np.copy(y1)
g2 = np.copy(y2)

g1[y1<0] = 0
g2[y2<0] = 0

a = -1*g1-1*g2

plt.plot(x,a)
```

```python

x = np.arange(-10,10,0.1)

y1 = -50-90*x
y2 = 100+50*x
y3 = 100+30*x

g1 = np.copy(y1)
g2 = np.copy(y2)
g3 = np.copy(y3)

g1[y1<0] = 0
g2[y2<0] = 0
g3[y3<0] = 0

a = g1+g2-g3

plt.plot(x,g1, 'r')
plt.plot(x,g2, 'g')
plt.plot(x,g3, 'b')

plt.plot(x,a,'k')
```
