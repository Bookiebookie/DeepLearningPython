 # What is Transfer Learning ?
 
Transfer learning is machine learning technique in which model trained for one particular task is used as a starting point for training model for another task. Transfer learning enable us to utilize knowledge ( like learned weights, features) from previously learned tasks and apply them to newer, related ones.  For example, in image classification task in deep learning,  certain low-level features such as edges, shapes, corners and intensity do not appears to be specific to particular dataset or particular task. Such learned features can be shared across different image classification task, thus enabling knowledge transfer among tasks. # Why Transfer Learning ?
Deep learning model are data-hungry, such that they need many training examples to learn the parameters in network. This is one of the limiting aspects of deep neural networks. And this is where the transfer learning comes in handy. Transfer learning is  common and highly effective approach to training the model on small image datasets by  utilising a **pre-trained models**.  Furthermore, transfer learning also helps in saving training time as new model is not trained from scratch. 

**What are pre-trained models ?**

Pre-trained models are deep learning architecture which are trained over large dataset. Such models are trained for thousands of hours for a given benchmark dataset and all the parameters in the networks are learned. There are numbers of such pre-trained models available as an open source project which anyone can use depending upon their problem. 

Few examples of pre-trained models are:

- **For Computer Vision tasks**
    - VGG-16
    - VGG-19
    - Inception V3
    - XCeption
    - ResNet-50
    - MobileNet

- **For Natural Language Processing task**
    - Word2Vec
    - GloVe
    - FastText
    - Universal Sentence Encoder by Google
    - Bidirectional Encoder Representations from Transformers (BERT) by Google


# Key considerations for Transfer Learning?
To effectively apply Transfer Learning we need to answer three main questions [3](https://www.cse.ust.hk/~qyang/Docs/2009/tkde_transfer_learning.pdf).
-  **What to transfer**: While applying transfer learning we need to understand what knowledge is common between the source and target task. We try to seek answer on what knowledge is specific to source task and what can be transferred from source task to target task that will help improve the performance of the target task. 
-  **When to transfer**: 'When to transfer' asks in which situation we should apply transfer learning. We should apply transfer learning only when source and target domain are related. When  source and target domains are not related at all, applying transfer learning degrades target task performance . This type of transfer is called Negative Transfer. 
- **How to transfer**: Once we have answer to what to transfer and when to transfer, we can proceed to identify the actual ways of transferring the knowledge across domains/tasks. 

**So, when does transfer learning makes sense?** 

If we are trying to learn from **Task A** and transfer knowledge to **Task B**, transfer learning is useful in following scenario [4](https://www.youtube.com/watch?v=yofjFQddwHE).
- Task A and B have same input x, e.g both have images as input. 
- There is alot more data for Task A than Task B.
- Low level features from A could be helpful in for learning task B.

# Transfer Learning Strategies
There are mainly two common ways pre-trained models are reused to train the new models from them <a href=https://towardsdatascience.com/a-comprehensive-hands-on-guide-to-transfer-learning-with-real-world-applications-in-deep-learning-212bf3b2f27a > [1] </a>. 

1. Using a pre-trained model as a feature extractor.
2. Fine-tuning the pre-trained model.

 ### Pre-trained Models as Feature Extractors
As we saw from notebook 3 that convolutional neural network consists of two parts: first part being the series of  convolutional and pooling layers and second part being densly connected layers. The first part is called 'convolutional base'<a href='https://github.com/fchollet/deep-learning-with-python-notebooks/blob/master/5.3-using-a-pretrained-convnet.ipynb'>[2].</a> The main idea in using pre-trained model as feature extractor is to remove fully connected layer of pre-trained model and use only convolutional base of it to extract the features from new images. And use these extracted features to train the new classifier for new task, which are created from randomly initialized the weights. In this method of transfer learning, we do not retrain pre-trained model or part of it(this is called freezing the network). The following figure demonstrates the concept of feature extraction from pre-trained model and then training a classifier on top of it. We have a pre-trained model trained on source data like ImageNet database. We chop of the densly connected layer in pre-trained model and use convolutional base to extract the feature from target data(i.e data for new classification task). And use the output from convolutional base to train the new model(target model) for new classification task in our hand.

###  Fine-tuning the pre-trained model 
In this process we do not just replace and retrain the classifier head, i.e fully connected layers, but we also selectively retrain some of the top layers used for feature extraction in pre-trained model. This is called "fine-tuning" because it slightly adjusts the more abstract representations of the model being reused, in order to make them more relevant for the problem at hand.

### Deciding which transfer learning strategy to use 
So, above we discussed different transfer learning strategies. But how do we decide which one to use? The decision to chose which  transfer learning strategy to use is based on mainly two factors: the size of the target (new) dataset (small or big), and its similarity to the source (original) dataset (e.g. ImageNet). We now discuss four common scenarios and how to navigate among them when choosing  transfer learning strategy.

**1. size of target dataset is small and similar to source dataset:** In this case, since size of the new dataset is low, fine tuning can lead to overfitting. And target data is similar to source data, we use the pretrained model as a feature extractor while changing the output layer according new classification task.

**2. Size of the target dataset is small and dissimilar to source data:** In this case  we fine tune last layers while freezing the top layers.



**3.Size of the target dataset is large and similar to source data:** In such scenario, the best option would be to fine tune  through the complete network of pre trained model with a small learning rate. Since, the target dataset is large, there is less concern of overfitting.


**4.Size of the target dataset is large and dissimilar to source data:** Since the target dataset is large, we can  train the neural network from scratch according to our data. However, in practice it is very often still beneficial to initialize with weights from a pretrained model.

```python
train_cats_dir = os.path.join(train_dir, 'cats')  # directory with our training cat pictures
train_dogs_dir = os.path.join(train_dir, 'dogs')  # directory with our training dog pictures
validation_cats_dir = os.path.join(validation_dir, 'cats')  # directory with our validation cat pictures
validation_dogs_dir = os.path.join(validation_dir, 'dogs')  # directory with our validation dog pictures
```

```python
# count the number of images 
num_cats_tr = len(os.listdir(train_cats_dir))
num_dogs_tr = len(os.listdir(train_dogs_dir))

num_cats_val = len(os.listdir(validation_cats_dir))
num_dogs_val = len(os.listdir(validation_dogs_dir))

total_train = num_cats_tr + num_dogs_tr
total_val = num_cats_val + num_dogs_val

print('total training cat images:', num_cats_tr)
print('total training dog images:', num_dogs_tr)

print('total validation cat images:', num_cats_val)
print('total validation dog images:', num_dogs_val)
print("--")
print("Total training images:", total_train)
print("Total validation images:", total_val)
```

## Baseline model 

We will first create  simple CNN model from scratch to classify cat vs dog. This  baseline model will establish a minimum model performance to which all of our other models can be compared. 

```python
batch_size = 128
epochs = 15
IMG_HEIGHT = 150
IMG_WIDTH = 150
```

```python
train_image_generator = ImageDataGenerator(rescale=1./255) # Generator for training data
validation_image_generator = ImageDataGenerator(rescale=1./255) # Generator for  validation data
```

```python
train_data_gen = train_image_generator.flow_from_directory(batch_size=batch_size,
                                                           directory=train_dir,
                                                           shuffle=True,
                                                           target_size=(IMG_HEIGHT, IMG_WIDTH),
                                                           class_mode='binary')
```

```python
val_data_gen = validation_image_generator.flow_from_directory(batch_size=batch_size,
                                                              directory=validation_dir,
                                                              target_size=(IMG_HEIGHT, IMG_WIDTH),
                                                              class_mode='binary')
```

Let's try our train data generator.

```python
sample_training_images, _ = next(train_data_gen)
```

```python
 # function to plot images 
def plotImages(images_arr):
    fig, axes = plt.subplots(1, 5, figsize=(20,20))
    axes = axes.flatten()
    for img, ax in zip( images_arr, axes):
        ax.imshow(img)
        ax.axis('off')
    plt.tight_layout()
    plt.show()
```

```python
# plot 5 images 
plotImages(sample_training_images[:5])
```

Now, let's define our model.

```python
model = Sequential([
    Conv2D(16, 3, padding='same', activation='relu', input_shape=(IMG_HEIGHT, IMG_WIDTH ,3)),
    MaxPooling2D(),
    Conv2D(32, 3, padding='same', activation='relu'),
    MaxPooling2D(),
    Conv2D(64, 3, padding='same', activation='relu'),
    MaxPooling2D(),
    Flatten(),
    Dense(512, activation='relu'),
    Dense(1)
])
```

```python
# compile the model 
model.compile(optimizer='adam',
              loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
model.summary()
```

Let's train the model. 

```python
%%time
history = model.fit(
    train_data_gen,
    steps_per_epoch=total_train // batch_size,
    epochs=epochs,
    validation_data=val_data_gen,
    validation_steps=total_val // batch_size
)
```

**Visualize training results**

```python
# function to plot model accuracy and loss in each epoch
def accuracy_loss_plot(history):
    
    '''function to plot model accuracy and loss in each epoch'''
    
    # Retrieve a list of list results on training and validation data
    # sets for each training epoch
    
    acc      = history.history['accuracy']
    val_acc  = history.history['val_accuracy']
    loss     = history.history['loss']
    val_loss = history.history['val_loss']
    epochs   = range(1,len(acc)+1) # Get number of epochs
    #------------------------------------------------
    # Plot training and validation accuracy per epoch
    #------------------------------------------------
    plt.figure(figsize=(10, 6))
    plt.subplot(1, 2, 1)
    plt.plot(epochs, acc,  label='Training accuracy')
    plt.plot(epochs, val_acc,  label='Validation accuracy')
    plt.title('Training and validation accuracy')
    plt.legend()
    
    plt.subplot(1, 2, 2)
    plt.plot(epochs, loss,  label='Training loss')
    plt.plot(epochs, val_loss,  label='Validation loss')
    plt.title('Training and validation loss')
    plt.legend()
```

```python
accuracy_loss_plot(history)
```

From above graph, we have achieved the training accuracy of 95% but validation accuracy is just around 70%. This means our model is clearly overfitting the training data.

# Pre-trained VGG16 model as a Feature Extractor 

There are two ways pre-trained model can be used as feature extractor <a href='https://github.com/fchollet/deep-learning-with-python-notebooks/blob/master/5.3-using-a-pretrained-convnet.ipynb'>[2]</a>. 
- One of the way is to use convolution base as a standalone feature extractor. In this setting, convolution base is used as preprocessing step and extract the features from the images. And use the extracted feature to train the separate classifier made up of dense layers. This method is very fast and cheap to run, because it only requires running the convolutional base once for every input image, and the convolutional base is by far the most expensive part of the pipeline <a href='https://github.com/fchollet/deep-learning-with-python-notebooks/blob/master/5.3-using-a-pretrained-convnet.ipynb'>[2]</a>. However, for the exact same reason, this technique would not allow us to leverage data augmentation at all. We call this method  feature extraction without image augmentation.


- Another way of using pre-trained as a feature extractor is by extending the model (convolutional base) by adding Dense layers on top, and running the whole thing end-to-end on the input data. This allows  to use data augmentation, because every input image is going through the convolutional base every time it is seen by the model. However, for this same reason, this technique is far more expensive than the first one. We call this method  feature extraction with image augmentation.

We will cover both of the above methods. 
Let's begin with first method.

###  Creating the base model from the pre-trained VGG16

First, load up the VGG-16 model without its densly connected layer. 

```python
conv_base = VGG16(weights='imagenet',
                  include_top=False,
                  input_shape=(IMG_HEIGHT, IMG_WIDTH ,  3))
```

In above block of code, we imported pre-trained VGG16 model from Keras. 

We passed three arguments to the constructor:

- **weights**, to specify which weight checkpoint to initialize the model from. We specified 'imagenet'. This means we imported VGG16 model with pre-trained weight derived from training over imagenet database. 
- **include_top**, which refers to include or not to include the densely-connected classifier on top of the network. We chose False. This means we chop densly connected classifier from VGG16 model and retained only its convolutional base. 
- **input_shape**, the shape of the image tensors that we will feed to the network.

```python
# let's print the summary of conv_base
conv_base.summary()
```

We can see from above summary result, the last activation feature map (output from block5_pool)  in  VGG-16 model gives us the  bottleneck features (output) of shape (4,4,512). These features map is then flattened and fed to our own fully connected deep neural network classifier to classify between cat and dog. Also, number of trainable parameter in convolutional base model is more than 14 million.

#### VGG16 model as feature extractor without image augmentation

Now, we are going to use convoluional base of VGG-16 model to extract the feature from our train and validation images. We adopted code from <a href='https://github.com/fchollet/deep-learning-with-python-notebooks/blob/master/5.3-using-a-pretrained-convnet.ipynb'>[2]</a>

```python
# function to extract the features from VGG16 model

def bottleneck_feature(data_dir, sample_count):

    # instantiate the data generator. 
    datagen = ImageDataGenerator(rescale= 1/255.0)
    batch_size = 32
    features = np.zeros(shape=(sample_count, 4, 4, 512))
    labels = np.zeros(shape=(sample_count))

    generator = datagen.flow_from_directory(
        data_dir,
        target_size=(150, 150),
        batch_size=batch_size,
        class_mode='binary')
    
    i = 0
    for inputs_batch, labels_batch in generator:
         #  here we generates output feature map  for the input samples.
        features_batch = conv_base.predict(inputs_batch)
        features[i * batch_size : (i + 1) * batch_size] = features_batch
        labels[i * batch_size : (i + 1) * batch_size] = labels_batch
        i += 1
        if i * batch_size >= sample_count:
            # Note that since generators yield data indefinitely in a loop,
            # we must `break` after every image has been seen once.
            break
    return features, labels
```

Extracting the bottleneck features.

```python
%%time
train_features, train_labels = bottleneck_feature(train_dir ,  total_train)
validation_features, validation_labels = bottleneck_feature(validation_dir, total_val)
```

```python
# Reshaping the output feature map 
train_features = np.reshape(train_features, (2000, 4 * 4 * 512))
validation_features = np.reshape(validation_features, (1000, 4 * 4 * 512))
```

```python
print(f"shape of training feature {train_features.shape}")
print(f"shape of validation feature {validation_features.shape}")
```

Now, let's define our own densely-connected classifier and train it on feature we extracted.

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(1,activation="sigmoid"))
```

```python
# compile the model 
from tensorflow.keras import optimizers
model.compile(optimizer=optimizers.RMSprop(lr=2e-5),
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

Train the model

```python
%%time
history = model.fit(train_features, train_labels,
                    epochs=30,
                    batch_size=20,
                    validation_data=(validation_features, validation_labels))
```

```python
accuracy_loss_plot(history)
```

Now We can see that using VGG-16 as standalone feature extractor we achieved a validation accuracy of about 87%, which much better than what we achieved in the previous section with our small model trained from scratch. However, our plots also indicate that model is overfitting the training data.

#### VGG-16 model as feature extractor with Image Augmentation

In this method, rather than using conv_base model from VGG-16 as standalone feature extractor, we add it to a Sequential model just like we add a layer. This method much slower and more expensive, but it allows us to leverage data augmentation during training. 

**Freezing the convolutional base**

In this step, we  freeze the conv_base model from the previous step as we want to use it  just as a feature extractor. Freezing the layers means avoiding the updates of weights from that layers during training the model.

<div class="alert alert-info">
    In Keras, freezing a network is done by setting its <b>trainable attribute to False</b>
</div> 

```python
conv_base.trainable = False
```

**Adding the classification head**

Now, we will stack the dense layer as classification layer on the top of feature extractor after flattening the feature map.


Final model 

```python
model = tf.keras.models.Sequential()

# here we add convolutional base as layer 
model.add(conv_base)

model.add(tf.keras.layers.Flatten()) 
model.add(tf.keras.layers.Dense(256, activation='relu'))
model.add(tf.keras.layers.Dense(1,activation="sigmoid"))
```

Let's see the result from summary function.

```python
model.summary()
```

From the result above we see now only the weights from the last layers that we added will be trained. All the weights associated with conv_base layer is now converted into Non-trainable parameter.

Let's now again configure train and validation generator. This time we use image augmentation while training.

```python
train_datagen = ImageDataGenerator(
      rescale=1./255,
      rotation_range=40,
      width_shift_range=0.2,
      height_shift_range=0.2,
      shear_range=0.2,
      zoom_range=0.2,
      horizontal_flip=True,
      fill_mode='nearest')

# Note that the validation data should not be augmented!
valid_datagen = ImageDataGenerator(rescale=1./255)



train_data_gen = train_datagen.flow_from_directory(batch_size=batch_size,
                                                           directory=train_dir,
                                                           shuffle=True,
                                                           target_size=(IMG_HEIGHT, IMG_WIDTH),
                                                           class_mode='binary')

val_data_gen = valid_datagen .flow_from_directory(batch_size=batch_size,
                                                              directory=validation_dir,
                                                              target_size=(IMG_HEIGHT, IMG_WIDTH),
                                                              class_mode='binary')
```

Compile the model 

```python
model.compile(loss='binary_crossentropy',
              optimizer=optimizers.RMSprop(lr=2e-5),
              metrics=['accuracy'])
```

Train the model

```python
%%time

history = model.fit(
    train_data_gen,
    steps_per_epoch=100,
    epochs=30,
    validation_data=val_data_gen,
    validation_steps=100
)
```

```python
accuracy_loss_plot(history)
```

In above figure we can see we are able to avoid overfitting unlike in feature extraction without image augmentation. At the mean time, validation accuracy has also slighly increased.

 #  Fine-tuning Pre-trained VGG-16 model

Earlier layers in the convolutional base encode more generic, reusable features like  texture, corners, edges and color blobs, while layers higher up encode more specialized features like  eye, nose, clothe item etc. One of the way to try to increase performance of the model  even further is to train (or "fine-tune") the weights of the top layers of the pre-trained model alongside the training of the classifier we added. This training process will force the weights to be tuned from generic feature maps to features associated specifically with the dataset which we are trying to train new model. However, we should be careful in choosing how many layers in pre-trained model we want to fine-tune. As we see earlier, pre-trained model has millions of parameters, so more parameters we are training, the more we are at risk of overfitting the training data especially if we have small dataset. 

Therefore, keeping this in mind, in our situation, it is a good strategy to only fine-tune the couple of layers in the convolutional base.

In our VGG-16 model object stored in the **conv_base** variable, we unfreeze last four layers.

**First we unfreeze our convolutional base model**

```python
conv_base.trainable = True
```

**Now, Freezing the all layers except last 4 layers**

```python
# Freezing the layers except the last 4 layers
for layer in conv_base.layers[:-4]:

    layer.trainable = False
```

```python
# Checking the trainable status of the individual layers

for layer in conv_base.layers:
    
    print(layer,"  ",  layer.trainable)
```

The true flag associated with layer means those layers are trainable. 

```python
conv_base.summary()
```

Since we unfreeze few layers in convolutional base,  we have much more learnable parameter in network than when using VGG16 just as a feature extractor. 

Note
<div class="alert alert-warning"> 
In order for these changes to take effect, we must first compile the model. If you ever modify weight trainability after compilation, you should then re-compile the model, or these changes would be ignored.

So, let's compile the model again. 

```python
model.compile(loss='binary_crossentropy',
              optimizer=tf.keras.optimizers.RMSprop(lr=1e-5),
              metrics=['accuracy'])
```

Now we can start fine-tuning our network. We will do this with the RMSprop optimizer, using a very low learning rate. The reason for using a low learning rate is that we want to limit the magnitude of the modifications we make to the representations of the 3 layers that we are fine-tuning. Updates that are too large may harm these representations.


```python
# we will fine tune the model for 25 epochs.
fine_tune_epochs = 50
initial_epochs = 30 # number of epoch we used to train the classifier earlier
total_epochs =  initial_epochs + fine_tune_epochs
```

Before starting fine tuning the model, let's understand **initial_epoch**. This  parameter specifies epoch at which to start training. This means we didn't randomly initialized the weight of our dense layer, rather update the weight from earlier training.  

```python
%%time
history_fine_tune = model.fit(
    train_data_gen,
    #steps_per_epoch=total_train // batch_size,
    steps_per_epoch=100,
    epochs=total_epochs,
    validation_data=val_data_gen,
    #validation_steps=total_val // batch_size,
    validation_steps=50,
    initial_epoch =  history.epoch[-1]
)
```

#### Plotting accuracy and loss 

```python
# first retrive the accuracy and loss from fine tuned model and append the result to result obtained from 
# earlier model
history.history['accuracy'] += history_fine_tune.history['accuracy']
history.history['val_accuracy'] += history_fine_tune.history['val_accuracy']

history.history['loss'] += history_fine_tune.history['loss']
history.history['val_loss'] += history_fine_tune.history['val_loss']
```

```python
#-----------------------------------------------------------
# Retrieve a list of list results on training and validation data
# sets for each training epoch and append the result to earlier accuracy results.
#-----------------------------------------------------------
acc      = history.history['accuracy']
val_acc  = history.history['val_accuracy']
loss     = history.history['loss']
val_loss = history.history['val_loss']
```

```python
plt.figure(figsize=(8, 8))
plt.subplot(2, 1, 1)
plt.plot(acc, label='Training Accuracy')
plt.plot(val_acc, label='Validation Accuracy')
#plt.ylim([0.8, 1])
plt.plot([initial_epochs-1,initial_epochs-1],
          plt.ylim(), label='Start Fine Tuning')
plt.legend(loc='lower center')
plt.title('Training and Validation Accuracy')

plt.subplot(2, 1, 2)
plt.plot(loss, label='Training Loss')
plt.plot(val_loss, label='Validation Loss')
#plt.ylim([0, 1.0])
plt.plot([initial_epochs-1,initial_epochs-1],
         plt.ylim(), label='Start Fine Tuning')
plt.legend(loc='lower left')
plt.title('Training and Validation Loss')
plt.xlabel('epoch')
plt.show()
```

We see that fine tuning has improved both the validation and train accuracy but we also have overfitting.

 # Summary
 
- We illustrates the advantage of transfer learning for smaller data-set applications. By reusing the pre-trained model we can get much better accuracy than with classifier trained from scratch.

- While reusing pre-trained model on a new dataset, via feature extraction , the pre-trained model is "frozen" and only the weights of the  newly classifier get updated during training. 

- In case of small data, feature extraction with image augmentation helps in avoiding overfitting but comes at the cost of higher computational power.

- Fine-tuning help in adjusting the weights that is more specific to  new dataset and can pushes performance of model a bit further.
